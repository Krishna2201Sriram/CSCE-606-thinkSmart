```python
#Import data required for 6th grade prediction
from pandas import read_excel
df2=read_excel("Corrected_6_Grade_HES - HJH STAAR Cohort Data for APLUS.xlsx",sheet_name="6th 2018-2019",skiprows=1)
df1=read_excel("Corrected_6_Grade_HES - HJH STAAR Cohort Data for APLUS.xlsx",sheet_name="6th Grade 2017-2018",skiprows=1)
```

```python
merged = df1.merge(df2, left_index=True, right_index=True,
                 how='outer', suffixes=('', '_y'))
merged.drop(merged.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)
```

```python
merged.head()
```

```python
merged=merged[merged['Spring 2018 STAAR\nMA04\nPcntScore\n5/2019'].notna()]
merged=merged[merged['Spring 2019 STAAR\nRE04\nPcntScore\n5/2019'].notna()]
```

```python
merged = merged[merged.isnull().sum(axis=1) < 7] 
merged.shape
```

```python
math=merged[['LocalId','Grade','Ethnicity','ECD - should correspond to Free Meal','LEP','SpEd','3rd math  Fall BM\n12/13/2017','3rd mathSpring BM\n3/29/2018','Spring 2018 STAAR\n5/2018\nMA03\nPcntScore','Fall 2018 Math BM\n12/18/2018','Spring 2018 Math BM\n02/13/2019','Spring 2018 STAAR\nMA04\nPcntScore\n5/2019']]
reading=merged[['LocalId','Grade','Ethnicity','ECD - should correspond to Free Meal','LEP','SpEd','Local\nEN.03.018.E\nPcntScore\n12/13/2017','Local\nEN.03.099.E\nPcntScore\n03/29/2018','Spring 2018 STAAR\nRE03\nPcntScore\n5/2018','Fall 2018 ELAR BM\n12/18/2018','Spring 2018 ELAR Released BM\n2/14/2019','Spring 2019 STAAR\nRE04\nPcntScore\n5/2019']]
```

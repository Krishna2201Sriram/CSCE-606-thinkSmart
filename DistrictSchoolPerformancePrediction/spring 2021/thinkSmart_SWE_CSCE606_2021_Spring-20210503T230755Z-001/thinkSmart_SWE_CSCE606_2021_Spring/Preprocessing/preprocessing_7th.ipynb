```python
#Import data required for 6th grade prediction
from pandas import read_excel
df1=read_excel("Corrected_6_Grade_HES - HJH STAAR Cohort Data for APLUS.xlsx",sheet_name="7th Grade 2018-2019",skiprows=1)
df2=read_excel("Corrected_6_Grade_HES - HJH STAAR Cohort Data for APLUS.xlsx",sheet_name="7th Grade 2017-2018",skiprows=1)
```

```python
merged = df1.merge(df2, left_index=True, right_index=True,
                 how='outer', suffixes=('', '_y'))
merged.drop(merged.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)
```

```python
merged=merged[merged['Spring 2019 STAAR\nMA05\nPcntScore\n5/2019 or 6/2019'].notna()]
merged=merged[merged['Spring 2019 STAAR\nRE05\nPcntScore 5/2019 or 6/2019'].notna()]
```

```python
merged = merged[merged.isnull().sum(axis=1) < 7] 
```

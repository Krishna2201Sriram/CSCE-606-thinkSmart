```python
#Import data required for 6th grade prediction
from pandas import read_excel
df1=read_excel("Corrected_6_Grade_HES - HJH STAAR Cohort Data for APLUS.xlsx",sheet_name="8th Grade 2018-2019",skiprows=1)
df2=read_excel("Corrected_6_Grade_HES - HJH STAAR Cohort Data for APLUS.xlsx",sheet_name="8th Grade 2017-2018",skiprows=1)
```

```python
df2.head()
```

```python
merged = df1.merge(df2, left_index=True, right_index=True,
                 how='outer', suffixes=('', '_y'))
merged.drop(merged.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)
```

```python
merged
```

```python
merged=merged[merged['Math 2019 STAAR Percent Score'].notna()]
merged=merged[merged['Reading 2019 STAAR Percent Score'].notna()]
```

```python
merged.shape
merged.filter['spring 2019 staar']
```

```python
#Remove students who's NAN values are more than 7
merged = merged[merged.isnull().sum(axis=1) < 7] 
```

```python
merged.shape
```

```python
merged.head()
```

```python
math=merged[['LocalId','Grade','Ethnicity','ECD','LEP','SpEd','Math Fall 2017 STAAR Practice Test Score','Math Spring 2018 STAAR Practice Test Score','Math Spring 2018 STAAR Percent Score \n','Math Fall 2018 STAAR Practice Test Percent Score','Math Spring 2019 STAAR Practice Test Percent Score','Math 2019 STAAR Percent Score']]
reading=merged[['LocalId','Grade','Ethnicity','ECD','LEP','SpEd','Reading Fall 2017 STAAR Practice Test Score','Reading Spring 2018 STAAR Practice Test Score\n','Reading Spring 2018 STAAR Percent Score \n','Reading Fall 2018 STAAR Practice Test Percent Score','Reading Spring 2019 STAAR Practice Test Percent Score','Reading 2019 STAAR Percent Score']]
```

```python
math.head()
reading.head()
```

```python
math=math.fillna(math.mean())
reading=reading.fillna(reading.mean())
```

```python
math.to_csv(r'math_8th.csv',index=False)
reading.to_csv(r'reading_8th.csv',index=False)
```
